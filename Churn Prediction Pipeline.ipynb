# Splitting the datasets

```python
X=df.drop('Churn',axis=1)
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

# Creating Pipeline

```python
from sklearn.preprocessing import MinMaxScaler
#from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
```

```python
#importing models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier 
```

```python
pipeline_lr=Pipeline([('scalr1',MinMaxScaler()),
                     ('lr_classifier',LogisticRegression(random_state=0))])
```

```python
pipeline_dt=Pipeline([('scalr2',MinMaxScaler()),
                     ('dt_classifier',DecisionTreeClassifier())])
```

```python
pipeline_rf=Pipeline([('scalr3',MinMaxScaler()),
                     ('rf_classifier',RandomForestClassifier())])
```

```python
pipeline_nb=Pipeline([('scalr4',MinMaxScaler()),
                     ('nb_classifier',GaussianNB())])
```

```python
pipeline_knn=Pipeline([('scalr5',MinMaxScaler()),
                     ('knn_classifier',KNeighborsClassifier())])
```

```python
pipelines=[pipeline_lr,pipeline_dt,pipeline_rf,pipeline_nb,pipeline_knn]
```

```python
best_accuracy=0.0
best_classifier=0
best_pipeline=''
```

```python
#pipeline dictionary
pipe_dict={0:'Logistic Regression',1:'Decission Tree',2:'Random Forest',3:'Naive_Bayes',4:'KNN'}
```

```python
for pipe in pipelines:
    pipe.fit(X_train,y_train)
```

```python
for i,model in enumerate(pipelines):
     print("{} Test Accuracy: {}".format(pipe_dict[i],model.score(X_test,y_test)))
```

```python
for i,model in enumerate(pipelines):
    if model.score(X_test,y_test)>best_accuracy:
        best_accuracy=model.score(X_test,y_test)
        best_pipeline=model
        best_classifier=i
print('Classifier with best accuracy:{}'.format(pipe_dict[best_classifier]))
```
